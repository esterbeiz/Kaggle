```python
newcols = set()
for el in dt['team_members']:
    if isinstance(el, dict):
        newcols.update(el.keys())
    elif isinstance(el, list):
        for i in el:
            newcols.update(i.keys())

def extractvalues(x, col):
    if isinstance(x['team_members'], dict):
        return x['team_members'].get(col, np.nan)
    elif isinstance(x['team_members'], list):
        return ','.join(str(i.get(col, '')) for i in x['team_members']).strip(',')

for nc in newcols:
    dt[nc] = dt.apply(lambda r: extractvalues(r, nc), axis=1)

dt.drop('team_members', axis=1, inplace=True)
```
