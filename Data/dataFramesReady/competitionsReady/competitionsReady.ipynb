```python
# Importacion de librerias

import pandas as pd
```

```python
# Lectura del fichero CompetitionsParsed
# Posicionarse en un ruta donde sea se pueda leer el fichero comodamente

%cd C:\Users\Usuario\Desktop\Kaggle\Data\dataFramesParsed\Competitions_Parsed

df_competitions = pd.read_csv("Dt_Competitions_Info_Raw.csv", index_col=0)
df_competitions.head()
```

Los principales cambios que se realizarán en este dataFrame son:
    - Renombrar las columnas con los nombres normalizados
    - Convertir teams a formato Int

#### Convertir teams a formato Int

```python
df_competitions['teams']=[each.replace("teams","") for each in df_competitions['teams']]
df_competitions['teams']=[each.replace(",","") for each in df_competitions['teams']]
df_competitions['teams'] = df_competitions['teams'].astype(float).astype(int)
```

```python
df_competitions.head()
```

```python
df_competitions.shape
```

```python
%cd C:\Users\Usuario\Desktop\Kaggle\Data\dataFramesApoyo

idCompetitions = pd.read_excel('id_competiciones_integrado.xlsx', index_col=0)
idCompetitions = idCompetitions[["competitionTitle","id_competitions"]]

idCompetitions.head()
```

```python
df_competitions = df_competitions.rename(columns={
                           'title':'competitionTitle'                       
                                 })

df_competitions = pd.merge(idCompetitions, df_competitions, how = "inner", on = "competitionTitle")
```

```python
df_competitions.shape
```

```python
df_competitions
```

#### Exportacion del dataFrame

```python
df_competitions = df_competitions[["date",
                                     "prize",
                                      "topic",
                                      "teams",
                                      "competitionTitle",
                                   "id_competitions"
                                      ]]
df_competitions = df_competitions.rename(columns={'date':'lastSubmission',
                          'prize':'rewardQuantity', "teams":"totalTeams","topic":"hostSegment",
                        "id_competitions": "competitionId"
                                 })
df_competitions.head()
```
