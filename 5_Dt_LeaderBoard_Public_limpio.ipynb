Cambiamos los datos 'SubmissionDate' a formato Date y Score a float.

```python
#Cambio Score
df.Score.unique()
```

```python
df["Score"] = df.Score.astype(float)
```

```python
#Cambio SubmissionDate
df['date'] = df['SubmissionDate'].str[:11]
df['time'] = df['SubmissionDate'].str[12:]
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df["title"] = df["title"].str.replace("-", " ").str.replace("publicleaderboard", "").str.replace(".csv","")
```

```python
df.head()
```

```python
df=df.drop(['SubmissionDate'], axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
g=pd.DataFrame(df.groupby('title')['TeamId'].nunique())
g.reset_index(level=0, inplace=True)
g.head()
```

```python
g["title"] = g["title"].str.strip()
```

```python
g = df.groupby('title')['TeamId'].value_counts().reset_index(name='entries')
g['total_entries_by_comp'] = df.groupby('title')['TeamId'].transform('count')
g.head()
```

```python
g.drop('total_entries_by_comp', axis=1, inplace=True)
g.head()
```

```python
df_epc = df[['title','entries']].groupby('entries').sum()
df_epc.reset_index(level=0, inplace=True)
df_epc.head()
```

```python
df['entries'] =1       #initially, set that counter to 1.
group_data = df.groupby(['title','TeamId'])['entries'].sum() #sum function
group_data.head()
```

```python
f=df.pivot_table('entries', ['title', 'TeamId], aggfunc='co').reset_index()
f.head()
```

```python
g = df.groupby('title')['TeamId'].value_counts().reset_index(name='entries')
g['total_e_by_comp'] = df.groupby('title')['TeamId'].transform('count')
```

```python
g.head()
```

```python
df_maxscorebyteam = df[['date', 'time','TeamId', 'TeamName','Score']].groupby('TeamId').max().reset_index()
df_maxscorebyteam.head()
```

```python
g.shape
```

```python
df_maxscorebyteam.shape
```

```python
h=pd.merge(g, df_maxscorebyteam, on='TeamId')
h.head()
```

```python
columns = ['title', 'TeamId', 'entries', 'total_entries_by_comp', 'date', 'time','TeamName', 'Score']
h = h.reindex(columns=columns)
```

```python
entries_per_team = h[['TeamName','entries']].groupby('TeamName').sum()
entries_per_team.reset_index(level=0, inplace=True)
entries_per_team.sort_values(by='entries', ascending=False)
```

```python
df_entriesteam = h[['TeamName','entries']].groupby('TeamName').sum()
df_entriesteam.reset_index(level=0, inplace=True)
df_entriesteam.sort_values(by='entries',ascending=False,inplace=True)
plt.figure(figsize=(30,15))
sns.barplot(x=h['TeamName'],y=h['entries'])
plt.title('Entries por equipo',fontsize = 20)
plt.xlabel('Nombre de Equipo',fontsize = 20)
plt.ylabel('Entries',fontsize = 20)
plt.xticks(rotation=45)
```
