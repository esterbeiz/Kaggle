Del dataframe g, seleccionamos las columnas 'title' y 'entries' para posteriormente realizar un 'join' con el dataframe que obtenemos de 'Dt_Competitions_INFO'.

Para ello realizamos un nuevo dataframe con 'title' y 'total_entries_by_comp':

```python
df_entries_per_comp = h[['title','entries']].groupby('title').sum()
df_entries_per_comp.reset_index(level=0, inplace=True)
df_entries_per_comp.head()
```

```python
df_entries_per_date = h[['title','TeamId']].groupby('title').count()
df_entries_per_date.reset_index(level=0, inplace=True)
df_entries_per_date.head()
```

```python
title_entries_team=pd.merge(df_entries_per_comp, df_entries_per_date, on='title')
title_entries_team.head()
```

```python
print(title_entries_team.isnull().sum())
```

```python
title_entries_team=title_entries_team.rename(columns = {'TeamId':'teams'})
title_entries_team.head()
```
